```python
import numpy as np
import matplotlib.pyplot as plt

# Number of iterations for the simulation
iterations = 10000

# Define the task durations (Optimistic, Pessimistic, Most Likely)
tasks = [
    [2, 3, 5],   # Task 1: User registration
    [1, 2, 4],   # Task 2: Payment method setup
    [1, 2, 3],   # Task 3: Reminder setup
    [3, 4, 6],   # Task 4: UI design and integration
    [3, 5, 7],   # Task 5: Database setup
    [2, 3, 5],   # Task 6: Payment history feature
    [1, 2, 4],   # Task 7: Payment reminder notifications
    [1, 2, 4],   # Task 8: Financial analysis graph
    [2, 3, 5],   # Task 9: Edit user settings
    [3, 4, 6],   # Task 10: Analyze financial habits
    [2, 3, 5],   # Task 11: Progress on financial goals
    [2, 3, 4],   # Task 12: Spontaneous spending analysis
]

# Monte Carlo Simulation based on tasks and durations
simulated_results = []

for i in range(iterations):
    total_time = 0

    # Randomly sample task durations using triangular distribution
    for task in tasks:
        # Ensure mode is correctly placed between optimistic and pessimistic
        if not (task[0] <= task[1] <= task[2]):
            task[0], task[1], task[2] = sorted(task)  # Sort to make sure mode is between optimistic and pessimistic

        total_time += np.random.triangular(task[0], task[1], task[2])

    simulated_results.append(total_time)

# Convert to numpy array for easier analysis
simulated_results = np.array(simulated_results)

# Calculate the 80% confidence interval
lower_bound = np.percentile(simulated_results, 10)  # 10th percentile (lower bound)
upper_bound = np.percentile(simulated_results, 90)  # 90th percentile (upper bound)

# Output the results
print(f"Прогнозируемое время выполнения Фазы 5 с 80% достоверностью: от {lower_bound:.2f} до {upper_bound:.2f} недель.")

# Plotting the distribution of the results
plt.hist(simulated_results, bins=50, alpha=0.7, color='blue', label="Simulation Results")
plt.axvline(x=lower_bound, color='red', linestyle='dashed', linewidth=2, label=f"10th percentile: {lower_bound:.2f}")
plt.axvline(x=upper_bound, color='green', linestyle='dashed', linewidth=2, label=f"90th percentile: {upper_bound:.2f}")
plt.xlabel('Weeks to Complete Phase 5')
plt.ylabel('Frequency')
plt.title('Monte Carlo Simulation for Phase 5 Completion')
plt.legend()
plt.show()
```
